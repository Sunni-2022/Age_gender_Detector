```python
#zooming an image
img=cv2.imread("C:\\Users\\SHAIK SABIHA\\OneDrive\\Desktop\\Age Gender Detector\\resources\\4.jpg")
(height,width)=img1.shape[0:2]
zoom=cv2.resize(img,None,fx=1.5,fy=1.5,interpolation=cv2.INTER_CUBIC)
plt.imshow(zoom)
```

```python
zoom=cv2.resize(img,(int(width*2),int(height*2)),interpolation=cv2.INTER_CUBIC)
plt.imshow(zoom)
```

```python
#rotating an image
m=cv2.getRotationMatrix2D((width/2,height/2),30,1)
rotate=cv2.warpAffine(img1,m,(height,width))
plt.imshow(rotate)
```
